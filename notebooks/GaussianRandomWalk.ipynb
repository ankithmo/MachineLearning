```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def do_walk(runs, steps):
    walks=[]
    for i in range(runs):
        X=[]
        X.append(0)
        for j in range(steps):
            X.append(X[j] + np.random.randn())
        walks.append(X)
    return walks
```

```python
walks = do_walk(3, 5000)
```

```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(1,1,1)
```

```python
for i in range(3):
    ax.plot(walks[i])
```

```python
plt.show()
```

```python
walks = do_walk(1000,5000)
```

```python
dist = np.zeros(1000)
for i in range(1000):
    dist[i] = walks[i][4999]
```

```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(1,1,1)
ax.hist(dist)
plt.show()
```

So we see that the final points of our random walks are distributed according to something that pretty much looks like a Gaussian distribution.
